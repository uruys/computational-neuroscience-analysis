```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

csv_path = "deeplabcut_mouse.csv"
output_dir = "figures"

os.makedirs(output_dir, exist_ok=True)

df = pd.read_csv(csv_path, header=[0,1,2])

scorer = df.columns.get_level_values(0)[0]
bodyparts = df[scorer].columns.get_level_values(0).unique()

summary = []

for bp in bodyparts:
    
    x = df[(scorer, bp, "x")].values
    y = df[(scorer, bp, "y")].values
    
    dx = np.diff(x, prepend=x[0])
    dy = np.diff(y, prepend=y[0])
    
    velocity = np.sqrt(dx**2 + dy**2)
    
    frames = np.arange(len(x))
    
    plt.figure()
    plt.plot(frames, x, label="x")
    plt.plot(frames, y, label="y")
    plt.title(bp + " position")
    plt.legend()
    plt.savefig(f"{output_dir}/{bp}_xy.png")
    plt.close()
    
    summary.append({
        "bodypart": bp,
        "mean_velocity": np.mean(velocity)
    })

summary_df = pd.DataFrame(summary)
summary_df.to_csv(f"{output_dir}/summary_metrics.csv", index=False)

print("Terminé")
```
